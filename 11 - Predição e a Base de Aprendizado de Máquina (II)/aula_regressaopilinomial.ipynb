```python
import numpy as np
import matplotlib.pyplot as plt  
```

```python
np.random.seed(2)
pageSpeeds = np.random.normal(3,1,1000)
purchaseAmount = np.random.normal(50,10,1000)/pageSpeeds

plt.scatter(pageSpeeds, purchaseAmount)
```

```python
x = np.array(pageSpeeds)
y = np.array(purchaseAmount)

p4 = np.poly1d(np.polyfit(x,y,4)) # cria um polinomio de 4Â° grau
```

```python
xp = np.linspace(0,7,100)

#porem uma aproximaÃ§Ã£o de 4Â° grau Ã© muito boa
plt.scatter(x,y)
plt.plot(xp,p4(xp), c='r')
plt.show()
```

```python
#fator R^2

from sklearn.metrics import r2_score

r2 = r2_score(y, p4(x))

print (r2)

```

```python
#caso fosse grau 2

p2 = np.poly1d(np.polyfit(x,y,2)) # cria um polinomio de 2Â° grau

xp = np.linspace(0,7,100)

plt.scatter(x,y)
plt.plot(xp,p2(xp), c='r')
plt.show()

r2 = r2_score(y, p4(x))

print (r2) # menos expressivo que o de 4Â° grau

# com essa funÃ§Ã£o (em vermelho), obtem-se uma relaÃ§Ã£o entre velocidade do rolar da pÃ¡gina e gastos com compras que pode ser extrapolada para supor novos dados
```
