```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import expon, poisson
```

#### Média moda e mediana

```python
dados = [10, 15, 20, 25, 25, 30, 35, 40, 40, 40, 45, 50, 55]
```

```python
media = np.mean(dados)
moda = stats.mode(dados)[0]
mediana = np.median(dados)

print(f'Média: {media}')
print(f'Moda: {moda}')
print(f'Mediana: {mediana}')
```

#### Variância e desvio padrão e distribuição normal

```python
dados = np.random.normal(100.0, 20.0, 10000)
plt.hist(dados, 50)
plt.show()
```

```python
dados.var()
```

```python
print(dados.std())
print(np.sqrt(dados.var()))
```

#### Distribuição uniforme

```python
valores = np.random.uniform(-10, 10, 100000)
plt.hist(valores, 50)
plt.show()
```

#### Distribuição exponencial

```python
dados = np.arange(0, 10, 0.0001)
plt.plot(dados, expon.pdf(dados))
```

#### Distribuição Poisson

```python
mu = 500
dados = np.arange(500, 600, 0.5)
plt.plot(dados, poisson.pmf(dados, mu))
```

#### Percentil 

```python

dados = [10, 15, 20, 25, 25, 30, 35, 40, 40, 40, 45, 50, 55]


percentil_25 = np.percentile(dados, 25)
percentil_50 = np.percentile(dados, 50) 
percentil_75 = np.percentile(dados, 75)

print(f'25º Percentil: {percentil_25}')
print(f'50º Percentil: {percentil_50}')
print(f'75º Percentil: {percentil_75}')
```
