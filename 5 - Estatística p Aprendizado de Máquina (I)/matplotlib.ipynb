```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
x = np.arange(-3,3,0.001)
print(x)
```

```python
plt.plot(x, norm.pdf(x))
plt.show()
```

```python
x = np.arange(-3,3,0.001)
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1, 0.5))
plt.show()
```

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1, 0.5))
plt.savefig("here.png", format='png')
```

```python
axes = plt.axes()
axes.set_xlim([-5,5])
axes.set_ylim([0,1])
plt.plot(x, norm.pdf(x))
plt.show()
```

```python
x = np.arange(-3,3,0.001)
plt.plot(x, norm.pdf(x), 'b-')
plt.plot(x, norm.pdf(x, 1, 0.5), 'r:')
plt.show()
```

```python
x = np.arange(-3,3,0.001)
plt.plot(x, norm.pdf(x), 'b-')
plt.plot(x, norm.pdf(x, 1, 0.5), 'r:')
plt.xlabel("legenda1")
plt.ylabel("legenda2")
plt.legend(['azul','laranja'], loc=4)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.xticks([])
plt.yticks([])
ax.set_ylim([-30,30])

data = np.ones(100)
data[70:] -= np.arange(30)

plt.annotate(
    "THE DAY I REALIZED...",
    xy=(70,1),
    arrowprops=dict(arrowstyle='->'),
    xytext=(15,-10)
)

plt.plot(data)
plt.xlabel("time")
plt.ylabel("my overall health")
```
