#### Lidando com outliers

Outliers são valores ou observações em um conjunto de dados que se distanciam significativamente dos outros valores. Esses pontos de dados são atípicos e podem indicar variabilidade extrema ou erros no conjunto de dados. 

```python
%matplotlib inline
import numpy as np

incomes = np.random.normal(27000, 15000, 10000)
incomes = np.append(incomes, [10000000000])

import matplotlib.pyplot as plt
plt.hist(incomes, 50)
plt.show()
```

```python
incomes.mean()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def reject_outliers(data):
    # mediana dos dados
    u = np.median(data)
    # desvio padrão dos dados
    s = np.std(data)
    # filtra os dados, mantendo apenas aqueles dentro de 2 desvios padrão da mediana
    filtered = [e for e in data if ((u - 2 * s) < e < (u + 2 * s))]

    return filtered


incomes = [50000, 60000, 65000, 70000, 80000, 1200000]  
filtered = reject_outliers(incomes)

plt.hist(filtered, 50)
plt.show()
```
