```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# gerando um conjunto de dados aleatórios com média 3, desvio padrão 1 e 1000 pontos de dados
pageSpeeds = np.random.normal(3.0, 1.0, 1000)


# cada purchaseAmount é calculado como 100 - (pageSpeeds + um pequeno ruído aleatório) * 3
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3

# gráfico de dispersão
plt.scatter(pageSpeeds, purchaseAmount)

```

```python
from scipy import stats
#regressão linear dos dados
slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
```

```python
#coeficiente de determinação (R^2) que indica a qualidade do ajuste da regressão
r_value ** 2 
```

```python
predict = lambda x : slope*x + intercept #função para prever valores com base na regressão

fitLine = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
```
