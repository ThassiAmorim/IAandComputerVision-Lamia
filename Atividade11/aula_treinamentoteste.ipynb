```python
%matplotlib inline
import numpy as np
from pylab import *

np.random.seed(2)

pageSpeeds = np.random.normal(3.0, 1.0, 100)

# gera 100 valores aleatórios com distribuição normal com média 50 e desvio padrão 30
# e depois divide cada valor pelo correspondente em 'pageSpeeds'
purchaseAmount = np.random.normal(50.0, 30.0, 100) / pageSpeeds

scatter(pageSpeeds, purchaseAmount)
```

```python
# divide os dados em conjuntos de 80% treinamento e 20% teste 
trainX = pageSpeeds[:80]
testX = pageSpeeds[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]
```

```python

scatter(trainX, trainY) #treinamento
```

```python

scatter(testX, testY) #teste
```

```python


x = np.array(trainX)
y = np.array(trainY)

# ajusta um polinômio de grau 8 aos dados de treinamento
p4 = np.poly1d(np.polyfit(x, y, 8))
```

```python
import matplotlib.pyplot as plt

#gera 100 pontos igualmente espaçados entre 0 e 7 para plotar a curva polinomial
xp = np.linspace(0, 7, 100)

axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])

plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python

testx = np.array(testX)
testy = np.array(testY)

axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])

# plota os pontos de teste e a curva do polinômio ajustado
plt.scatter(testx, testy)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

#  coeficiente R^2 para teste
r2 = r2_score(testy, p4(testx))
print(r2)
```

```python

# coeficiente R^2 para os dados de treinamento
r2 = r2_score(np.array(trainY), p4(np.array(trainX)))
print(r2)
```
